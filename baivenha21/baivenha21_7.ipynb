```python
import math

x1= float(input("Hoành độ điểm 1: "))
y1= float(input("Tung độ điểm 1: "))
x2= float(input("Hoành độ điểm 2: "))
y2= float(input("Tung độ điểm 2: "))
x3= float(input("Hoành độ điểm 3: "))
y3= float(input("Tung độ điểm 3: "))
x4= float(input("Hoành độ điểm 4: "))
y4= float(input("Tung độ điểm 4: "))

AB = BA = math.sqrt((x1-x2)**2+(y1-y2)**2)
BC = CB = math.sqrt((x2-x3)**2+(y2-y3)**2)
CD = DC = math.sqrt((x3-x4)**2+(y3-y4)**2)
AD = DA = math.sqrt((x1-x4)**2+(y1-y4)**2)
BD = DB = math.sqrt((x2-x4)**2+(y2-y4)**2)
AC = CA = math.sqrt((x1-x3)**2+(y1-y3)**2)


cosA = (AB**2 + AD**2 - BD**2)/(2*AB*AD)
cosB = ((AB**2 + BC**2 - AC**2)/(2*AB*BC))
cosC = ((BC**2 + CD**2 - BD**2)/(2*CD*BC))
cosD = ((AD**2 + CD**2 - AC**2)/(2*AD*DC))

#Khúc này cẩn thận hơn thì có thể xét điều kiện cosA,cosB,cosC,cosD luôn

gA=(math.degrees(math.acos(cosA)))
gB=(math.degrees(math.acos(cosB)))
gC=(math.degrees(math.acos(cosC)))
gD=(math.degrees(math.acos(cosD)))

if not round(gA+gB+gC+gD) == 360:
    print("Đây không là tứ giác")
else:                                   #Khúc này mình định cho nó nhảy lên lại để nhập lại toạ độ mà hổng bik lệnh 
    if gA <= 180 and gB <= 180 and gC <= 180 and gD <= 180 :
            print("Đây là tứ giác lồi")
    else :
        print("Đây là tứ giác lõm")

       

import turtle
t = turtle.Turtle()

t.color('blue')
t.pensize(4)
t.penup()
t.goto(x1,y1)
t.pendown()
t.fillcolor('green')
t.begin_fill()
t.goto(x2,y2)
t.goto(x3,y3)
t.goto(x4,y4)
t.goto(x1,y1)
t.end_fill()
t.hideturtle()

turtle.done()










```
